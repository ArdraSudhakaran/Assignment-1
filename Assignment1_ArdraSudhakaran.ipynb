5..Perform meaningful visualizations using the data set.Bring atleast 3 visualizations

a.Histogram

```python
# Histogram using matplotlib

plt.figure(figsize=(8,5))
plt.hist(iris['PL'])
plt.title('Distribution of petal length')
plt.xlabel('Petal length')
plt.ylabel('Frequency')
plt.show()
```

```python
# Histogram using seaborn

sns.histplot(x=iris['SL'])
plt.title('Distribution of sepal length')
plt.show()
```

b. Boxplot

```python
plt.figure(figsize=(8,5))
sns.boxplot(x=iris['Classification'],y=iris['PW'])
plt.title('Boxplot of species data')
plt.show()
```

c. Scatter plot

```python
# Scatter plot of sepal length vs sepal width of 3 species


sns.scatterplot(x=iris['SL'],y=iris['SW'],hue=iris['Classification'])
plt.title('Scatter plot')
```

d. Pair plot

```python
# Pairplot to analyze the relationship between species for all characteristic combinations

sns.pairplot(iris,hue='Classification')
plt.title('Pairplot')
plt.show()
```
